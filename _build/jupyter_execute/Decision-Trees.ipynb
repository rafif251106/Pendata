# Decision Tree

## Konsep Dasar Decision Tree

Pohon keputusan atau decision tree merupakan algoritma machine learning yang fungsinya memecah dataset yang kompleks menjadi himpunan bagian yang lebih kecil dengan terus membangun struktur yang hierarkis.

Konsep dasar decision tree sebagai berikut:
1. Struktur Anatomi Pohon

    Decision Tree terdiri dari 3 elemen utama:
    * Akar

        Akar (Root Node) merupakan simpul paling atas yang mewakili seluruh populasi atau data. Titik ini merupakan titik awal di mana pemisahan pertama dilakukan.
    * Simpul Keputusan

        Simpul keputusan (Internal Node) merupakan titik di mana data dipisahkan berdasarkan fitur atau atribut tertentu. Setiap simpul memiliki cabang yang mewakili hasil dari pengujian tersebut.
    *  Daun 

        Daun (Leaf Node) merupakan Simpul akhir yang mewakili hasil prediksi. Di titik ini, tidak ada lagi percabangan yang dilakukan.

2. Cara Kerja

    Decision Tree bekerja dengan cara melakukan Recursive Partitioning atau pembagian berulang. Algoritma akan memilih atribut terbaik untuk membelah data menjadi kelompok yang sehomogen mungkin dengan memiliki karakteristik yang sama.

    Untuk menentukan atribut mana yang paling bagus sebagai pemisah, algoritma biasanya menggunakan metrik statistik:

    * Entrophy

        Entrophy merupakan Ukuran ketidakteraturan atau ketidakmurnian dalam sekumpulan data.

        Rumus entrophy:

        $$Entropy(S) = \sum_{i=1}^{n} -p_i \log_2 p_i$$

    * Information Gain

        Information gain merupakan Pengurangan nilai Entropy setelah data dipisahkan. Atribut dengan Information Gain tertinggi akan dipilih sebagai pemisah.

        Rumus Information Gain:

        $$Gain(S, A) = Entropy(S) - (\sum_{j=1}^{k} \frac{|S_j|}{|S|} \times Entropy(S_j))$$
    
    * SplitInfo

        SplitInfo digunakan untuk mengukur seberapa luas dan seragam data terpecah berdasarkan suatu atribut. Jika sebuah atribut membagi data menjadi banyak potongan kecil, nilai SplitInfo-nya akan tinggi.

        Rumus SplitInfo:

        $$SplitInfo_A(D) = -\sum_{j=1}^{v} \frac{|D_j|}{|D|} \times \log_2\left(\frac{|D_j|}{|D|}\right)$$
    
    * Gain Ratio

        Gain Ratio adalah hasil normalisasi dari Information Gain menggunakan SplitInfo. Metrik ini dipilih untuk menentukan atribut pemisah terbaik pada algoritma seperti C4.5.

        Rumus Gain Ratio:

        $$GainRatio(A) = \frac{Gain(A)}{SplitInfo_A(D)}$$

### Langkah-langkah Perhitungan Gain Ratio

1. Menghitung Entrophy Total

    Sebelum menghitung entrophy, Hitung banyak data kolom target yaitu kolom Beli dan kolom lainnya

    Kolom Beli:

    |Ya|Tidak|Total|
    |--|--|--|
    |13|7|20|

    Kolom Harga:

    | |Ya|Tidak|Total|
    |-|--|-----|-----|
    |Mahal|4|3|7|
    |Sedang|3|3|6|
    |Murah|6|1|7|

    Kolom Review

    | |Ya|Tidak|Total|
    |-|--|-----|-----|
    |Bagus|8|0|8|
    |Sedang|5|2|7|
    |Buruk|0|5|5|

    Kolom Diskon

    | |Ya|Tidak|Total|
    |-|--|-----|-----|
    |Ada|9|2|11|
    |Tidak|4|5|9|

    * Menghitung Entrophy Beli(Utama)

        $Entropy(Beli) = (-(\frac{13}{20}) \log_2(\frac{13}{20}))+(-(\frac{7}{20})\log_2(\frac{7}{20})) = 0,934$
    
    * Menghitung Entrophy Harga

        * Entrophy Mahal

            $Entropy(Mahal) = (-(\frac{4}{7}) \log_2(\frac{4}{7}))+(-(\frac{3}{7})\log_2(\frac{3}{7})) = 0,985$
        
        * Entrophy Sedang

            $Entropy(Sedang) = (-(\frac{3}{6}) \log_2(\frac{3}{6}))+(-(\frac{3}{6})\log_2(\frac{3}{6})) = 1$
        
        * Entrophy Murah

            $Entropy(Murah) = (-(\frac{6}{7}) \log_2(\frac{6}{7}))+(-(\frac{1}{7})\log_2(\frac{1}{7})) = 0,592$
        
        * Entrophy Harga

            $Entropy(Harga) = (\frac{7}{20} \times 0,985)+(\frac{6}{20} \times 1)+(\frac{7}{20} \times 0,592) = 0,852$
    
    * Menghitung Entrophy Review

        * Entrophy Bagus

            $Entropy(Bagus) = (-(\frac{8}{8}) \log_2(\frac{8}{8})) + (-(\frac{0}{8}) \log_2(\frac{0}{8})) = 0$

        * Entrophy Sedang

            $Entropy(Sedang) = (-(\frac{5}{7}) \log_2(\frac{5}{7})) + (-(\frac{2}{7}) \log_2(\frac{2}{7})) = 0,863$

        * Entrophy Buruk

            $Entropy(Buruk) = (-(\frac{0}{5}) \log_2(\frac{0}{5})) + (-(\frac{5}{5}) \log_2(\frac{5}{5})) = 0$

        * Entrophy Review

            $Entropy(Review) = (\frac{8}{20} \times 0) + (\frac{7}{20} \times 0,863) + (\frac{5}{20} \times 0) = 0,302$

    * Menghitung Entrophy Diskon
    
        * Entrophy Ada
    
            $Entropy(Ada) = (-(\frac{9}{11}) \log_2(\frac{9}{11})) + (-(\frac{2}{11}) \log_2(\frac{2}{11})) = 0,684$
    
        * Entrophy Tidak
    
            $Entropy(Tidak) = (-(\frac{4}{9}) \log_2(\frac{4}{9})) + (-(\frac{5}{9}) \log_2(\frac{5}{9})) = 0,991$
    
        * Entrophy Diskon
    
            $Entropy(Diskon) = (\frac{11}{20} \times 0,684) + (\frac{9}{20} \times 0,991) = 0,822$

2. Menghitung Information Gain

    * Information Gain Harga

        $Gain(Beli, Harga) = 0,934 - 0,852 = 0,082 $
    
    * Information Gain Review

        $Gain(Beli, Review) = 0,934 - 0,302 = 0,632 $
    
    * Information Gain Diskon

        $Gain(Beli, Harga) = 0,934 - 0,822 = 0,112 $

3. Menghitung Split Information

    * Split Information Harga

        $SplitInfo(Harga) = (-(\frac{7}{20}) \log_2(\frac{7}{20})) + (-(\frac{6}{20}) \log_2(\frac{6}{20})) + (-(\frac{7}{20}) \log_2(\frac{7}{20})) = 1,581 $
    
    * Split Information Review
        
        $SplitInfo(Review) = (-(\frac{8}{20}) \log_2(\frac{8}{20})) + (-(\frac{7}{20}) \log_2(\frac{7}{20})) + (-(\frac{5}{20}) \log_2(\frac{5}{20})) = 1,559 $
    
    * Split Information Diskon

        $SplitInfo(Diskon) = (-(\frac{11}{20}) \log_2(\frac{11}{20})) + (-(\frac{9}{20}) \log_2(\frac{9}{20})) = 0,933 $

4. Menghitung Gain Ratio

    * Gain Ratio Harga

        $GainRatio(Harga) = \frac{0,082}{1,581} = 0,052$
    
    * Gain Ratio Review

        $GainRatio(Review) = \frac{0,632}{1,559} = 0,405$
    
    * Gain Ratio Diskon

        $GainRatio(Diskon) = \frac{0,112}{0,933} = 0,113$

Dari Hasil Gain Ratio harga, review, dan diskon dapat dilihat bahwa Gain Ratio review paling tinggi, maka review akan menjadi Akar (Root Node)
