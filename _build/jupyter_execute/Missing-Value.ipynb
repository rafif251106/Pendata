# Normalisasi

Normalisasi adalah salah satu teknik persiapan data yang paling sering digunakan. Pada  machine learning dan  data mining, proses ini membantu kita mengubah nilai kolom numerik dalam dataset untuk menggunakan skala umum.
Tujuan utama dari normalisasi data yakni menghilangkan redundansi data (pengulangan) dan menstandarisasi informasi untuk alur kerja data yang lebih baik. Pembelajaran Mesin & Kecerdasan Buatan

Normalisasi data digunakan untuk menskalakan data suatu atribut sehingga berada dalam rentang yang lebih kecil, seperti -1 hingga 1 atau 0 hingga 1. Hal ini umumnya berguna untuk algoritma klasifikasi. 

Ada 3 Macam Normalisasi, yaitu:


## Min-Max Normalization

Min–Max Normalisasi adalah metode untuk mengubah nilai data numerik agar berada dalam rentang tertentu, biasanya 0 sampai 1 dengan menggunakan nilai terendah(min) dan nilai tertinggi(max). Tujuannya supaya skala antar fitur menjadi sama dan tidak ada nilai yang terlalu besar dibanding yang lain.

Rumus Min-Max Normalisasi:

$x' = \frac{x - x_{min}}{x_{max} - x_{min}}$

Contoh Min-Max Normalisasi:

| data |
|--------|
| 20 |
| 30 |
| 40 |

Min = 20, Max = 40

data ke-1
$x' = \frac{20 - 20}{40 - 20} = 0$

data ke-2
$x' = \frac{30 - 20}{40 - 20} = 0.5$

data ke-3
$x' = \frac{40 - 20}{40 - 20} = 1$

|data|normalisasi|
|----|-----------|
|20|0|
|30|0.5|
|40|1|

```python
def min_max(data):
    normalisasi = []
    angka_min = min(data)
    angka_max = max(data)
    for i in range(len(data)):
        jumlah = (data[i] - angka_min) / (angka_max - angka_min)
        normalisasi.append(jumlah)
    return normalisasi

data = [20,30,40]
min_max(data)
```

## Z-score Normalization

Z-score normalisasi (atau standardisasi) adalah metode normalisasi data yang mengubah nilai sehingga memiliki rata-rata (mean) = 0 dan simpangan baku (standar deviasi) = 1.
Metode ini menunjukkan seberapa jauh suatu nilai menyimpang dari rata-rata data dalam satuan standar deviasi.
Contoh interpretasi:

* Z = 0 → nilai sama dengan rata-rata
* Z = 1 → nilai 1 standar deviasi di atas rata-rata
* Z = −1 → nilai 1 standar deviasi di bawah rata-rata

Rumus Z-score Normalisasi:

$z = \frac{x - \mu}{\sigma}$

Keterangan:
* x = nilai data
* μ = rata-rata (mean) data
* σ = standar deviasi
* z = nilai hasil normalisasi (z-score)

Cara Menghitung Z-core Normalisasi:

|data|
|---|
|20|
|40|
|50|
|70|

* Menghitung Rata-Rata

$\mu = \frac{20 + 40 + 50 + 70}{4} = 45$

* Menghitung Standar Deviasi

$\sigma = \sqrt{\frac{(20-45)^2 + (40-45)^2 + (50-45)^2 + (70-45)^2}{4}} = \sqrt{\frac{625 + 25 + 25 + 625}{4}} = \sqrt{\frac{1300}{4}} = \sqrt{325} = 18.03$

* Menghitung Z-score

   * data ke-1

   $z = \frac{20 - 45}{18.03} = -1.39$

   * data ke-2

   $z = \frac{40 - 45}{18.03} = -0.28$

   * data ke-3

   $z = \frac{50 - 45}{18.03} = 0.28$

   * data ke-4

   $z = \frac{70 - 45}{18.03} = 1.39$

* Hasil Z-score Normalisasi

    |data|Normalisasi|
    |---|---|
    |20|-1.39|
    |40|-0.28|
    |50|0.28|
    |70|1.39|

```python
import math

def z_score(data):
    normalisasi = []
    deviasi = 0
    mean = sum(data) / len(data)
    for i in range(len(data)):
        deviasi += (data[i] - mean) ** 2
    
    std_dev = math.sqrt(deviasi / len(data))

    for i in range(len(data)):
        z = (data[i] - mean) / std_dev
        z = round(z, 2)
        normalisasi.append(z)
    
    return normalisasi

data = [20,40,50,70]
z_score(data)

```

## Decimal Scaling Normalization

Decimal Scaling Normalization adalah metode normalisasi data dengan cara menggeser posisi desimal pada nilai data sampai semua nilai berada dalam rentang tertentu (biasanya antara −1 sampai 1).

Rumus Decimal Scaling Normalisasi:

$x' = \frac{x}{10^j}$

Keterangan:
* x = nilai data asli
* x′= nilai setelah normalisasi
* 𝑗 = jumlah digit dari nilai terbesar (atau jumlah pergeseran desimal)

Cara Decimal Scaling Normalisasi:

|data|
|---|
|30|
|80|
|350|
|200|
|760|

* Tentukan angka terbesar dalam kolom data

Angka terbesar dalam data adalah 760, Karena 760 memiliki 3 digit angka maka j = 3, berarti rumus menjadi:

$x' = \frac{x}{10^3}$

* menghitung Decimal Scalling Normalization

    * data ke-1

    $x' = \frac{30}{10^3} = 0.03$

    * data ke-2

    $x' = \frac{80}{10^3} = 0.08$

    * data ke-3

    $x' = \frac{350}{10^3} = 0.35$

    * data ke-4

    $x' = \frac{200}{10^3} = 0.2$

    * data ke-5

    $x' = \frac{200}{10^3} = 0.76$

* Hasil Decimal Scaling Normalisasi

    |data|Normalisasi|
    |---|----|
    |30|0.03|
    |80|0.08|
    |350|0.35|
    |200|0.2|
    |760|0.76|

```python
def decimal_scaling(data):
    normalisasi = []
    angka_max = max(data)
    j = len(str(int(angka_max)))
    for i in range(len(data)):
        normalisasi.append(data[i] / (10 ** j))
    return normalisasi

data = [30,80,350,200,760]
decimal_scaling(data)
```

### Menggunakan Code Python

#### Dataset

```python
import pandas as pd
import numpy as np

data = {
    'umur': [25, 30, 22, 35, 28],
    'pendapatan': [3000, 4000, 2500, 5000, 3500],
    'skor': [60, np.nan, 55, 70, 65]
}

data_sementara = {
    'umur': [25,22, 35, 28],
    'pendapatan': [3000,2500, 5000, 3500],
    'skor': [60,55, 70, 65]
}

pd.DataFrame(data)
```

#### Menghitung Jarak

```python
import math
import pandas as pd

skor = []
umur = 30
pendapatan = 4000
for i in data['skor']:
    if not np.isnan(i):
        skor.append(i)

hasil_jarak = []
for i in range(len(data['umur'])):
    if np.isnan(data['skor'][i]):
        continue
    
    u = data['umur'][i]
    p = data['pendapatan'][i]
    
    jarak = math.sqrt((u - umur)**2 + (p - pendapatan)**2)
    
    hasil_jarak.append((round(jarak, 2)))



data_sementara["jarak"] = hasil_jarak
pd.DataFrame(data_sementara, columns=['umur', 'pendapatan', 'skor', 'jarak'])
```

#### Menghitung Bobot

```python
import pandas as pd

bobot = []
for i in data_sementara["jarak"]:
    bagi = 1 / i
    bobot.append(round(bagi, 5))

data_sementara["bobot"] = bobot

pd.DataFrame(data_sementara)
```

#### Menghitung WKNN

```python
import pandas as pd
from zmq import has

total_bobot = sum(data_sementara["bobot"])
jumlah = 0

for i in range(len(data_sementara["bobot"])):
    jumlah += data_sementara["skor"][i] * data_sementara["bobot"][i]

hasil_akhir = jumlah / total_bobot
hasil_akhir = round(hasil_akhir)

print("Hasil Prediksi Skor:", hasil_akhir)
print("")
print("Data Hasil Prediksi:")
data["skor"][1] = hasil_akhir
pd.DataFrame(data)
```
