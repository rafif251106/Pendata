## Analisa Data Kesuburan Tanah

### Pemrosesan Data menggunakan KNN

#### A. Preprocessing atau pemrosesan data

Data sebelum diolah harus dilakukan pemrosesan data supaya data yang digunakan telah bersih, pemrosesan yang saya lakukan diantaranya:

1. Missing Value

Pada dataset kesuburan tanah terdapat banyak sekali missing value jadi pada KNIME saya menggunakan widget missing value untuk memberikan nilai pada data missing value. Metode yang saya gunakan adalah:
* Number(float) menggunakan Mean
* String menggunakan Most Frequence Value

2. Tabel Partisioner
