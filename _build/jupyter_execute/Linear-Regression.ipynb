## Analisis Menggunakan Linear Regression

### Pengertian Linear Regression

Linear Regression (Regresi Linear) adalah metode statistik dan machine learning yang digunakan untuk mencari hubungan antara variabel input (X) dan variabel output (Y) dengan membentuk garis lurus terbaik. Seperti banyak ambisi manusia, idenya sederhana: tarik garis, lalu berharap realitas cukup sopan untuk mengikutinya.

### Hitung Manual

#### Rumus OLS


$$ \hat{\beta} = (X^{T}X)^{-1}X^{T}Y $$


1. Matriks variabel independen ($X$) dan vektor target ($Y$):

$$
X = \begin{bmatrix}
1 & 2 \\
1 & 4 \\
1 & 3 \\
1 & 3 \\
1 & 3 \\
1 & 4 \\
1 & 5
\end{bmatrix}, \quad
Y = \begin{bmatrix}
2 \\
3 \\
5 \\
4 \\
3 \\
5 \\
6
\end{bmatrix}
$$

2. Transpose dari matriks $X$:

$$
X^T = \begin{bmatrix}
1 & 1 & 1 & 1 & 1 & 1 & 1 \\
2 & 4 & 3 & 3 & 3 & 4 & 5
\end{bmatrix}
$$


3. Perkalian matriks $X^T$ dengan $X$:

$$
X^T X = \begin{bmatrix}
7 & 24 \\
24 & 88
\end{bmatrix}
$$

4. Mencari nilai Determinan dari $X^T X$:

$$ \det(X^T X) = (7)(88) - (24)(24) $$
$$ \det(X^T X) = 616 - 576 = 40 $$

5. Mencari matriks invers $(X^T X)^{-1}$:

$$ (X^T X)^{-1} = \frac{1}{40} \begin{bmatrix} 88 & -24 \\ -24 & 7 \end{bmatrix} $$
$$ (X^T X)^{-1} = \begin{bmatrix} 2.2 & -0.6 \\ -0.6 & 0.175 \end{bmatrix} $$


6. Perkalian matriks $X^T$ dengan $Y$:

$$
X^T Y = \begin{bmatrix}
28 \\
102
\end{bmatrix}
$$

7. Menghitung estimasi koefisien regresi ($\hat{\beta}$):

$$ \hat{\beta} = (X^T X)^{-1} X^T Y $$
$$ \hat{\beta} = \begin{bmatrix} 2.2 & -0.6 \\ -0.6 & 0.175 \end{bmatrix} \begin{bmatrix} 28 \\ 102 \end{bmatrix} $$
$$ \hat{\beta} = \begin{bmatrix} (2.2)(28) + (-0.6)(102) \\ (-0.6)(28) + (0.175)(102) \end{bmatrix} $$
$$ \hat{\beta} = \begin{bmatrix} 61.6 - 61.2 \\ -16.8 + 17.85 \end{bmatrix} $$
$$ \hat{\beta} = \begin{bmatrix} 0.4 \\ 1.05 \end{bmatrix} $$


8. Dari hasil di atas, kita mendapatkan nilai konstanta dan slope:

$$ \beta_0 = 0.4 $$
$$ \beta_1 = 1.05 $$

9. Sehingga persamaan model regresi linearnya adalah:

$$ \hat{Y} = 1.05X + 0.4 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data X (variabel independen)
x = np.array([2, 4, 3, 3, 3, 4, 5])

# Data Y (variabel dependen)
y = np.array([2, 3, 5, 4, 3, 5, 6])

# Ubah X menjadi matriks 2D (format yang diminta scikit-learn)
X = x.reshape(-1, 1)

# Membuat model Linear Regression
model = LinearRegression()

# Melatih model
model.fit(X, y)

# Mengambil koefisien regresi
beta_0 = model.intercept_   # Intercept (β0)
beta_1 = model.coef_[0]     # Slope (β1)


# Menampilkan persamaan regresi
print(f"\nPersamaan regresi:")
print(f"Y = {beta_0:.4f} + {beta_1:.4f}X")

# Prediksi nilai Y
y_pred = model.predict(X)

# Nilai R^2
r2 = model.score(X, y)

# ---------------------------
# Membuat grafik
# ---------------------------

# Scatter plot data asli
plt.scatter(x, y, label='Data Asli', s=80)

# Urutkan data agar garis regresi terlihat rapi
idx = np.argsort(x)
x_sorted = x[idx]
y_pred_sorted = y_pred[idx]

# Plot garis regresi
plt.plot(x_sorted, y_pred_sorted, linewidth=2,
         label=f'Garis Regresi: Y = {beta_0:.2f} + {beta_1:.2f}X')

# Judul dan label
plt.title('Linear Regression dengan Scikit-Learn')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Menampilkan grafik
plt.show()
```
