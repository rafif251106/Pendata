## Analisa Data Kesuburan Tanah

### Pemrosesan Data menggunakan KNN

#### A. Preprocessing atau pemrosesan data

Data sebelum diolah harus dilakukan pemrosesan data supaya data yang digunakan telah bersih, pemrosesan yang saya lakukan diantaranya:

1. Missing Value

Pada dataset kesuburan tanah terdapat banyak sekali missing value jadi pada KNIME saya menggunakan widget missing value untuk memberikan nilai pada data missing value. Metode yang saya gunakan adalah:
* Number(float) menggunakan Mean
* String menggunakan Most Frequence Value

Dataset Sebelum Misiing value

![Sebelum Missing Value](image/sebelum_missing.png)
Dataset Setelah Missing value

![Setelah Missing Value](image/setelah_missing.png)

2. Tabel Partisioner

Membagi data menjadi 2 dengan rasio 70% dan 30%

3. Normalisasi

Pada dataset kesuburan tanah saya menggunakan normalisasi min-max.
Hasil Normalisasi:

![Hasil Normalisasi](image/normalisasi.png)

4. PCA

PCA digunakan untuk memngubah dataset dengan fitur banyak menjadi sedikit dan agar dapat digunakan pada KNN, saya mengubah 10 fitur menjadi 4 fitur saja dengan nama PCA dimension 0 - 3.
Hasil PCA:

![Hasil PCA](image/pca.png)



#### B. Penerapan Algoritma KNN

Analisis yang saya lakukan yaitu melakukan algoritma KNN dengan data normalisasi lansung tanpa PCA dan Normalisasi dengan PCA.

Analisis KNN dengan normalisasi lansung tanpa PCA:

![Hasil KNN tanpa PCA](image/tanpa_pca.png)

Analisis KNN dengan normalisasi dengan PCA:

Dari kedua analisis tersebut terlihat bahwa
|Metrik|Nilai|
|------|-----|
|Accuracy|overall = 1 atau 100%|
|Precision|subur dan tidak subur = 1 atau 100%|
|Recall|subur dan tidak subur = 1 atau 100%|
|F1-Score|subur dan tidak subur = 1 atau 100%|
