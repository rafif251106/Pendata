```python
import csv
from itertools import count

from zmq import has

def read_csv(file_path):
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file,delimiter=';')
        data = [row for row in csv_reader]
    return data

def mean(data,colom):
    hasil = 0
    count = 0
    for row in data[1:]:
        hasil += float(row[colom])
        count += 1
    
    hasil /= count
    hasil = round(hasil,3)
    return hasil

def mode(data,colom):
    count = {}
    for row in data[1:]:
        nilai = row[colom]
        if nilai in count:
            count[nilai] += 1
        else:
            count[nilai] = 1
    
    hasil = 0
    value_awal = 0
    for key,value in count.items():
        if value > value_awal:
            value_awal = value
            hasil = key    
    return hasil 

def median(data,colom):
    nilai = len(data[1:])
    if nilai % 2 == 0:
        median1 = float(data[nilai // 2][colom])  
        median2  = float(data[nilai // 2 - 1][colom])
        hasil = (median1 + median2) / 2
    else:
        hasil = float(data[nilai // 2][colom])
    hasil = round(hasil,1)
    return hasil

def max(data,colom):
    hasil = 0
    for row in data[1:]:
        nilai = float(row[colom])
        if nilai > hasil:
            hasil = nilai
    return hasil

def min(data,colom):
    hasil = float(data[1][colom])
    for row in data[1:]:
        nilai = float(row[colom])
        if nilai < hasil:
            hasil = nilai
    return hasil

data = read_csv(r'C:\Users\asus\tugas\IRIS(2).csv')
```

## Mean


```python
for i in range(0,len(data[0])-1):
    print(f"Mean dari {data[0][i]} = {mean(data,i)}")
```

## Mode / Modus

```python
for i in range(0,len(data[0])-1):
    print(f"Modus dari {data[0][i]} = {mode(data,i)}")
```

## Median


```python
for i in range(0,len(data[0])-1):
    print(f"Median dari {data[0][i]} = {median(data,i)}")
```

## Max

```python
for i in range(0,len(data[0])-1):
    print(f"Max dari {data[0][i]} = {max(data,i)}")
```

## Min

```python
for i in range(0,len(data[0])-1):
    print(f"Min dari {data[0][i]} = {min(data,i)}")
```
